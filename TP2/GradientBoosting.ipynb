# Gradient Boosting

```python
import pandas as pd
train = pd.read_csv('prep_train.csv').filter(items=['Reason for absence', 'Distance from Residence to Work', 'Age', 'Work load Average/day ', 'Hit target', 'Disciplinary failure', 'Weight',  'Absent'])
train.head(10)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

param_grid = {
    'learning_rate'         : [0.01, 0.5, 0.1, 0.2, 1, 2],
    'loss'                  : ['deviance', 'exponential'],
    'min_samples_split'     : [2, 4, 8, 16, 32],
    'max_depth'             : [3, 10, 20, None],
    'min_impurity_decrease' : [ 0.1, 0.001, 0.00001, 0],
    'max_features'          : ['auto', 'log2', None]
    
}


X_train, X_test, y_train, y_test = train_test_split(train.drop(['Absent'],axis=1), 
                                                    train['Absent'], test_size=0.30, 
                                                    random_state=20)

grid = GridSearchCV(GradientBoostingClassifier(),param_grid,refit=True,verbose=3, scoring='balanced_accuracy')
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
predictions = grid.predict(X_test)
print(confusion_matrix(y_test,predictions))
print(accuracy_score(y_test,predictions))
```
