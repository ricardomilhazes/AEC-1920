# Support Vector Machines

```python
import pandas as pd
train = pd.read_csv('prep_train.csv')
train.head()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score



X_train, X_test, y_train, y_test = train_test_split(train.drop(['Absent'],axis=1), 
                                                    train['Absent'], test_size=0.30, 
                                                    random_state=1)
param_grid = {'gamma': ['scale', 'auto'],  \
              'tol': [ 0.0001, 0.001, 0.01], 'C': [ 0.1, 1, 10], \
              'shrinking': [True, False], \
              'kernel': ['rbf', 'linear', 'poly', 'sigmoid'], \
              'decision_function_shape': ['ovr', 'ovo'],  \
              'probability' : [True, False] }



grid1 = GridSearchCV(SVC(),param_grid,refit=True,verbose=3, scoring='balanced_accuracy')
grid1.fit(X_train,y_train)
```

```python
print(grid1.best_params_)
predictions = grid1.predict(X_test)
print(confusion_matrix(y_test,predictions))
print(accuracy_score(y_test,predictions))
```
