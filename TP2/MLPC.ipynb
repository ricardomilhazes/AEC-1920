```python
import pandas as pd
train = pd.read_csv('prep_train.csv')
train.head()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score



param_grid = {
    'hidden_layer_sizes'    : [(32, 16), (64, 32), (8,4)],
    'activation'            : ['identity', 'logistic', 'tanh', 'relu'],
    'solver'                : ['lbfgs', 'sgd', 'adam'],
    'alpha'                 : [0.0001, 0.001, 0.01],
    'learning_rate_init'    : [ 0.01, 0.001, 0.0001],
    'max_iter'              : [ 100, 200, 500]
}

X_train, X_test, y_train, y_test = train_test_split(train.drop(['Absent'],axis=1), 
                                                    train['Absent'], test_size=0.30, 
                                                    random_state=69)

grid = GridSearchCV(MLPClassifier(),param_grid,refit=True,verbose=3, scoring='balanced_accuracy')
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
predictions = grid.predict(X_test)
print(confusion_matrix(y_test,predictions))
print(accuracy_score(y_test,predictions))
```
